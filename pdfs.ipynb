```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Incubation period drawn from Weibull distribution (Figure 2B in paper)

from scipy.stats import weibull_min

fig, ax = plt.subplots(1, 1)
weibull_c = 2.322737 # shape
weibull_lambda = 6.492272 # scale
x = np.linspace(weibull_min.pdf(0.01, weibull_c, scale=weibull_lambda),
                weibull_min.ppf(0.99, weibull_c, scale=weibull_lambda), 100)
ax.plot(x, weibull_min.pdf(x, weibull_c, scale=weibull_lambda),'r-')
print("mean, variance:")
weibull_min.stats(weibull_c, scale=weibull_lambda)
```

```python
transmission_before_symptoms_percentage = 15
```

```python
# Skew-normal distribution for generating serial intervals (Figure 2C in paper)

from scipy.stats import skewnorm
fig, ax = plt.subplots(1, 1)
loc = 5
scale = 2
match transmission_before_symptoms_percentage:
            case 1:
                a = 30
            case 15:
                a = 1.95
            case 30:
                a = 0.7
x = np.linspace(skewnorm.ppf(0.01, a, loc=loc, scale=scale),
                skewnorm.ppf(0.99, a, loc=loc, scale=scale), 100)
ax.plot(x, skewnorm.pdf(x, a, loc=loc, scale=scale),
       'r-', label='skewnorm pdf')
skewnorm.cdf(loc, a, loc=loc, scale=scale)
```

```python
onset_to_isolation = "short"
```

```python
# Onset to isolation delay drawn from Weibull distribution (Figure 2A in paper)

from scipy.stats import weibull_min

fig, ax = plt.subplots(1, 1)
match onset_to_isolation:
            case "short":
                weibull_c = 1.651524
                weibull_lambda = 4.287786
            case "long":
                weibull_c = 2.305172
                weibull_lambda = 9.483875
x = np.linspace(weibull_min.pdf(0.01, weibull_c, scale=weibull_lambda),
                weibull_min.ppf(0.99, weibull_c, scale=weibull_lambda), 100)
ax.plot(x, weibull_min.pdf(x, weibull_c, scale=weibull_lambda),'r-')
print("mean, variance:")
weibull_min.stats(weibull_c, scale=weibull_lambda)
```

```python
# Negative binomial distribution for drawing number of new infections

from scipy.stats import nbinom

R_0 = 2.5
R_0_disp = 0.16
p = R_0_disp / (R_0_disp + R_0)
print(f"p={p}")

fig, ax = plt.subplots(1, 1)
x = np.arange(nbinom.ppf(0.01, n=R_0_disp, p=p),
              nbinom.ppf(0.99, n=R_0_disp, p=p))
ax.vlines(x, 0, nbinom.pmf(x, n=R_0_disp, p=p), lw=7)
```
